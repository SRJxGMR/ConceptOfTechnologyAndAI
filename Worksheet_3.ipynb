```python
def categorize_age(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

tic["age_group"] = tic["Age"].apply(categorize_age)
tic.head()
```

```python
total = len(tic)
total
```

```python
group_counts = tic["age_group"].value_counts()
group_counts
```

```python
group_percentage = (group_counts / total) * 100
group_percentage
```
